```python
# State
```

```python
df.columns
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.barplot(x = 'State', y = 'Orders', data = sales_state)
plt.show()
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(5) # checking top 5 states

sns.set(rc={'figure.figsize':(15,5)})  # adding chart size, width 15 and height 5, can be changed as suitable
sns.barplot(x = 'State', y = 'Orders', data = sales_state)
plt.show()

# total amount spent per state
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5) # checking top 5 states

sns.set(rc={'figure.figsize':(15,5)})  # adding chart size, width 15 and height 5, can be changed as suitable
sns.barplot(x = 'State', y = 'Amount', data = sales_state)
plt.show()

# total amount spent for state/sales
```

```python
# Marital Status
```

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(2,1)}) # readjusting the chart size

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5) # checking top 5 states

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x = 'Marital_Status', y = 'Amount', data = sales_state, hue = 'Gender')
plt.show()
```

```python
# Occupation
```

```python
sns.set(rc={'figure.figsize':(15,5)})

ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()

# which sector contributes most for sales?
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5) 

sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(x = 'Occupation', y = 'Amount', data = sales_state)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(20,5)})

ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5) 

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x = 'Product_Category', y = 'Amount', data = sales_state)
plt.show()

# most amount of money spent on which product?
```

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(5) 

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x = 'Product_ID', y = 'Orders', data = sales_state)
plt.show()
```
